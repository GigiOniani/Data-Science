```python
### Diabetes Prediction Project 

import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
data = pd.read_csv('C://Users/User/Desktop/diabetes.csv')
df= pd.DataFrame(data)
```

```python
#Lets split datasets into X for features and y for target
X = df.iloc[:,:8]
y= df.iloc[:,8:]
```

```python
#Check for possible N/A and nulls in given data
X.isna().sum()
X.isnull().sum()
```

```python
#Lets Check and review summary of our data 
X.describe()
correlation = X.corr()
sns.heatmap(correlation,annot=True,cmap='cool')
pd.plotting.scatter_matrix(X,y,figsize=(15,15));
    
```

```python
#split data for training and test validation sets
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=22)
```

```python
#Train the model with LogisticRegression module 
model = LogisticRegression()
clf= model.fit(X_train,y_train)
score = clf.score(X_train,y_train)
score_test = clf.score(X_test,y_test)

print(f'''
    model score for training instances is {round(score,2)}
    model score for test instances is {round(score_test,2)}''')
```

```python
#0.77 score Not bad at all,Lets play with another modules and try Decisionforests
from sklearn.tree import DecisionTreeClassifier
model2= DecisionTreeClassifier()
clf2= model2.fit(X_train,y_train)
score2=clf2.score(X_train,y_train)
score_test2=clf2.score(X_test,y_test)

print(f'''
    model score for test instances are {score_test2}''')
```
